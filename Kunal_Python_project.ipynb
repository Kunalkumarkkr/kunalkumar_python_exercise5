```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
np.random.seed(0)
n = 10                 #  Assign number of line segment
D = []
Theata = 45            # Assign angle by user
x1,y1 = 0,0            # Define a point(Origin) in a blue point

# identify points: a in blue, b in red
a = np.random.uniform(1, 10, (n, 2))
b = np.random.uniform(1, 10, (n, 2))
fig, ax = plt.subplots(figsize=(6,6))
ab_pairs = np.c_[a, b]
ab_args = ab_pairs.reshape(-1, 2, 2).swapaxes(1, 2).reshape(-1, 2)

# segments
ax.plot(*ab_args, c='k')

# identify points: a in blue, b in red
plt.plot(x1,y1 ,marker = 'o')
ax.plot(*a.T, 'bo')
ax.plot(*b.T, 'ro')
plt.show()

# Define function to identify shortest distance between a point and line segment
# Origin is define here as a x1,y1
# Given angle is Theata to a line segment direction 
# Formula to find out angle between a point on a line segment and origin is tanθ = y2-y1/x2-x1
# Put the value of x1,y1 in formula so x2tanθ = y2
# Define function to get a line equation ix2+by2+c = 0
# Using x2tanθ = y2 and equation ix2+by2+c = 0 , We get the value of x2,y2
# Using two point distance formula sqrt((x2-x1)**2 + (y2-y1)**2))
def coefficients(x,y): 
    global i,j,x2,y2
    coefficients = np.polyfit(x, y, 1)
    i = float(coefficients[0])
    j = float(coefficients[1])
    x2 = -1/(i+j*math.tan(Theata))
    y2 = x2*(math.tan(Theata))
for x,y in zip(a,b):
        coefficients(x,y)
        D.append(math.sqrt(abs((x2-x1)**2 + (y2-y1)**2))) 
Dist = min(D)
print('Shortest Distance is ' + str(Dist))
```
